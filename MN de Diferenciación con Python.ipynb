```python
import numpy as np
import matplotlib.pyplot as plt
from numpy import *
from pylab import plot,show
plt.style.use("seaborn-poster")
```

```python
def func(x):
    f=(2*x**2)+3*x
    return f
def fun_p(x):
    fp=4*x+3
    return fp
```

```python
h=0.1
x0=1.5
```

```python
fp_ad=(func(x0+h)-func(x0))/h
fp_at=(func(x0)-func(x0-h))/h
fp_ct=(func(x0+h)-func(x0-h))/(2*h)
print(fp_ad)
print(fp_at)
print(fp_ct)
```

```python
fpm_ct=(func(x0-2*h)-8*func(x0-h)+8*func(x0+h)-func(x0+2*h))/(12*h)
print(fpm_ct)
```

```python
x=linspace(-2,4,150)
rt=func(x0)+fp_ct*(x-x0)
plot(x,func(x),x0,func(x0),x,rt)
```
